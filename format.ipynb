```python
from ctypes import *
from ctypes.util import find_library
import pandas as pd
import numpy as np
```

```python


cols = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
df = pd.read_csv('abalone.csv', names = cols)

sex = {'M': 1, "F": -1, "I": 0}
df['Sex'] = df['Sex'].apply(lambda x: sex[x])
df['Rings'] = np.where(df['Rings']>=9, 1, -1)

df.head()
```

```python
cols_order = ['Rings','Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight']
df = df[cols_order]
df.head()
```

```python
labels = [str(i)+': ' for i in range(1,9)]

def labeling(x):
    data = x.values
    st = str(data[0]) + ' '
    for i in range(1,9):
        st = st+ labels[i-1] + str(data[i])+' '
    return st

# Sample
a = df.iloc[0]
labeling(a)
```

```python
data_set = df.apply(lambda x: labeling(x), axis = 1)
data_set
```

```python
data_set = df.apply(lambda x: labeling(x), axis = 1)
data_set[:3133].to_csv('train.data',index = False, header = False)
data_set[3133:].to_csv('test.data',index = False, header = False)
```
